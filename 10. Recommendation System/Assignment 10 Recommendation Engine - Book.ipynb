```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
book = pd.read_csv(r"D:\DS\10. Recommendation System\book.csv", encoding="latin1")
book.head()
```

```python
book.shape
```

```python
# dropping unnecessary column

book = book.drop(columns='Unnamed: 0', axis=1)
book.head()
```

```python
# renaming the columns

book.rename(columns={'User.ID':'UserID', 'Book.Title':'BookTitle', 'Book.Rating':'BookRating'}, inplace=True)
book.head()
```

```python
# sorting the table by userid and reseting the index column

book.sort_values('UserID', ascending=True).reset_index(drop=True)
```

```python
# number of unique users 

len(book.UserID.unique())
```

```python
# number of unique books

len(book.BookTitle.unique())
```

```python
# creating pivot and imputing NaN qith 0

book2 = book.pivot_table(values='BookRating', index='UserID', columns='BookTitle')
book2.fillna(0, inplace=True)
book2.head()
```

```python
# Importing Libraries

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
user_sim = 1- pairwise_distances(book2.values, metric='cosine')
user_sim
```

```python
# storing the result in dataframe format

user_sim2 = pd.DataFrame(user_sim)
user_sim2.head()
```

```python
# Set the index and column names to user ids 

user_sim2.index = book['UserID'].unique()
user_sim2.columns=book['UserID'].unique()
user_sim2
```

```python
user_sim2.iloc[:5,:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim2.iloc[:5,:5]
```

```python
# most similar user

user_sim2.idxmax(axis=1)[0:5]
```
